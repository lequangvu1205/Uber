```python
# Đếm số dòng trong file
import csv
import io

# Đọc raw text trước
with open(r"C:\Users\THE PCCUSTOM\Downloads\uber\uber_rapido_driver_dirty.csv", 
          encoding="utf-8-sig") as f:
    raw = f.read()

# Strip outer quotes của từng dòng rồi parse lại
lines = []
for line in raw.splitlines():
    line = line.strip()
    # Nếu dòng bắt đầu và kết thúc bằng " thì strip nó ra
    if line.startswith('"') and line.endswith('"'):
        line = line[1:-1]
    # Thay "" (escaped quote bên trong) thành "
    line = line.replace('""', '"')
    lines.append(line)

# Ghép lại thành string và đọc bằng pandas
cleaned = "\n".join(lines)
df = pd.read_csv(io.StringIO(cleaned))

print(df.shape)
df.head()
```

```python
# Khám phá sơ bộ (EDA nhanh)
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
df.describe(include='all')
```

```python
#Drop duplicates
df = df.drop_duplicates()
print("Sau drop duplicates:", df.shape)
```

```python
# Chuẩn hóa placeholder về Nan
import numpy as np

placeholders = ['-', 'n/a', 'N/A', 'na', 'none', 'None', 'unknown', 'Unknown', 'NULL', 'null', '']
df = df.replace(placeholders, np.nan)
```

```python
# Xử lí dữ liệu từng cột bị xấu
# age: strip "yrs", loại outlier
df['age'] = df['age'].str.replace(r'\D', '', regex=True)
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df.loc[~df['age'].between(18, 70), 'age'] = np.nan

# avg_daily_earnings_INR: strip ký tự thừa
df['avg_daily_earnings_INR'] = df['avg_daily_earnings_INR'].str.replace(r'[^\d.]', '', regex=True)
df['avg_daily_earnings_INR'] = pd.to_numeric(df['avg_daily_earnings_INR'], errors='coerce')

# Chuẩn hóa boolean
bool_true  = ['yes','YES','Yes','true','True','TRUE','1','Y','y']
bool_false = ['no','NO','No','false','False','FALSE','0','N','n']
for col in ['is_currently_active', 'peak_hour_availability']:
    df[col] = df[col].apply(lambda x: 1 if x in bool_true else (0 if x in bool_false else np.nan))

# Chuẩn hóa categorical (strip + title case)
for col in ['city', 'platform', 'vehicle_type', 'vehicle_condition',
            'background_check_status', 'document_verification', 'interview_result']:
    df[col] = df[col].str.strip().str.title()

# Scores > 10 → NaN (out of range)
score_cols = ['communication_score_10','driving_test_score_10','route_knowledge_score_10',
              'customer_handling_score_10','overall_interview_score_10']
for col in score_cols:
    df.loc[df[col] > 10, col] = np.nan

# Rating > 5 → NaN
df.loc[pd.to_numeric(df['overall_driver_rating'], errors='coerce') > 5, 'overall_driver_rating'] = np.nan
df['overall_driver_rating'] = pd.to_numeric(df['overall_driver_rating'], errors='coerce')
```

```python
# Chuẩn hóa ngày tháng
for col in ['interview_date', 'onboarding_date']:
    df[col] = pd.to_datetime(df[col], dayfirst=False, errors='coerce')
```
